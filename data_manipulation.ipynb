<b>Final comparison of the models</b>

```python
from sklearn.model_selection import KFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
import pandas as pd
import numpy as np

df = pd.read_csv('titanic.csv')
df['male'] = df['Sex'] == 'male'

kf = KFold(n_splits=5, shuffle=True)

X1 = df[['Pclass', 'male', 'Age', 'Siblings/Spouses', 'Parents/Children', 'Fare']].values
X2 = df[['Pclass', 'male', 'Age']].values
X3 = df[['Fare', 'Age']].values
y = df['Survived'].values

def model_scores(X, y, kf):
    accuracy_scores = []
    precision_scores = []
    recall_scores = []
    f1_scores = []
    for train_i, test_i in kf.split(X):
        X_train, X_test = X[train_i], X[test_i]
        y_train, y_test = y[train_i], y[test_i]
        model = LogisticRegression()
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        accuracy_scores.append(model.score(X_test, y_test))
        precision_scores.append(precision_score(y_test, y_pred))
        recall_scores.append(recall_score(y_test, y_pred))
        f1_scores.append(f1_score(y_test, y_pred))
    print('Accuracy score:', np.mean(accuracy_scores))
    print('Precision score:', np.mean(precision_scores))
    print('Recall score:', np.mean(recall_scores))
    print('F1 score:', np.mean(f1_scores))

print('Logistic regression with all features')
model_scores(X1, y, kf)
print()
print('Logistic regression with Pclass, male, age')
model_scores(X2, y, kf)
print()
print('Logistic regression with Fare and age')
model_scores(X3, y, kf)
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=10)

def score(kf, X, y):
    dt_accuracy_score = []
    dt_recall_score = []
    dt_precision_score = []
    lr_accuracy_score = []
    lr_recall_score = []
    lr_precision_score = []
    
    for tr_i, tt_i in kf.split(X):
        X_train, X_test = X[tr_i], X[tt_i]
        y_train, y_test = y[tr_i], y[tt_i]
        dt = DecisionTreeClassifier()
        dt.fit(X_train, y_train)
        dt_y_pred = dt.predict(X_test)
        dt_accuracy_score.append(dt.score(X_test, y_test))
        dt_precision_score.append(precision_score(y_test, dt_y_pred))
        dt_recall_score.append(recall_score(y_test, dt_y_pred))
        lr = LogisticRegression()
        lr.fit(X_train, y_train)
        lr_y_pred = lr.predict(X_test)
        lr_accuracy_score.append(lr.score(X_test, y_test))
        lr_precision_score.append(precision_score(y_test, lr_y_pred))
        lr_recall_score.append(recall_score(y_test, lr_y_pred))
        
    print('Decision tree')
    print('accuracy:', np.mean(dt_accuracy_score))
    print('precision:', np.mean(dt_precision_score))
    print('recall:', np.mean(dt_recall_score))
    print('Logistic regression')
    print('accuracy:', np.mean(lr_accuracy_score))
    print('precision:', np.mean(lr_precision_score))
    print('recall:', np.mean(lr_recall_score))
score(kf, X, y)
        
    
```

criterion of the Decision tree

```python
kf = KFold(n_splits=5, shuffle=True)
for criterion in ['gini', 'entropy']:
    print("Decision Tree - {}".format(criterion))
    accuracy = []
    precision = []
    recall = []
    for train_index, test_index in kf.split(X):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        dt = DecisionTreeClassifier(criterion=criterion)
        dt.fit(X_train, y_train)
        y_pred = dt.predict(X_test)
        accuracy.append(accuracy_score(y_test, y_pred))
        precision.append(precision_score(y_test, y_pred))
        recall.append(recall_score(y_test, y_pred))
    print("accuracy:", np.mean(accuracy))
    print("precision:", np.mean(precision))
    print("recall:", np.mean(recall), '\n')
    print()
```

```python
y_pred = mlp.predict(X_test)
incorrect = X_test[y_pred != y_test]
incorrect_true = y_test[y_pred != y_test]
incorrect_pred = y_pred[y_pred != y_test] 
```
